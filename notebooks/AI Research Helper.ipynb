```python
from llama_index.core import SimpleDirectoryReader

reader = SimpleDirectoryReader(input_dir="/kaggle/input/air-quality-research-papers")
documents = reader.load_data()
```

```python
from llama_index.core import Document
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core.ingestion import IngestionPipeline

# Create pipeline
pipeline = IngestionPipeline(
    transformations=[
        SentenceSplitter(chunk_size=512, chunk_overlap=64),  # recommended settings
        HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5"),
    ]
)

# Run the pipeline on your loaded documents (from previous step)
nodes = pipeline.run(documents=documents)
```

```python
print(len(nodes))
print(nodes[0].text[:500])  # Preview text of the first chunk
print(nodes[0].metadata)
```

```python
import chromadb
from llama_index.vector_stores.chroma import ChromaVectorStore

chroma_client = chromadb.PersistentClient(path="/kaggle/working/chroma_db")
chroma_collection = chroma_client.get_or_create_collection("air_quality_papers")
vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
vector_store.add(nodes)
```

```python
# Retrieve raw documents from the collection
docs = chroma_collection.get(include=["metadatas", "documents"], limit=3)
for doc, meta in zip(docs["documents"], docs["metadatas"]):
    print("---")
    print("TEXT:", doc[:300])
    print("META:", meta)
```

```python
print("Number of items in Chroma collection:", len(chroma_collection.get()["ids"]))
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
```

```python
from llama_index.core import VectorStoreIndex

index = VectorStoreIndex.from_vector_store(
    vector_store=vector_store,
    embed_model=embed_model
)
```

```python
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
token= user_secrets.get_secret("HF_TOKEN")
```

```python
from llama_index.llms.huggingface_api import HuggingFaceInferenceAPI

llm = HuggingFaceInferenceAPI(
    model_name="Qwen/Qwen2.5-Coder-32B-Instruct",
    token=token
)

query_engine = index.as_query_engine(
    llm=llm,
    response_mode="tree_summarize",
)

response = query_engine.query("What are the recent machine learning techniques for air quality prediction?")
print(response)
```

```python
from llama_index.core.agent.workflow import AgentWorkflow
from llama_index.llms.huggingface_api import HuggingFaceInferenceAPI
from llama_index.core.tools import QueryEngineTool

# Step 1: Initialize the LLM
llm = HuggingFaceInferenceAPI(
    model_name="Qwen/Qwen2.5-Coder-32B-Instruct",
    token=token
)

# Step 2: Create QueryEngineTool (already done, repeat only if needed)
query_engine_tool = QueryEngineTool.from_defaults(
    query_engine=query_engine,
    name="AirQualityResearchRAG",
    description="Use this tool to answer questions about air quality prediction from research papers.",
    return_direct=False  # optional; set to True to skip reasoning
)

# Step 3: Wrap in AgentWorkflow
agent = AgentWorkflow.from_tools_or_functions(
    [query_engine_tool],
    llm=llm,
    system_prompt=(
        "You are a research assistant that helps answer technical questions about air quality prediction. "
        "Use the tools provided to access information from academic papers and respond clearly and accurately."
    )
)
```

```python
response = await agent.run("What deep learning models are commonly used for AQI prediction?")
print(response)
```

```python
from llama_index.core.evaluation import FaithfulnessEvaluator

# Initialize evaluators
faithfulness_evaluator = FaithfulnessEvaluator(llm=llm)


query = "What deep learning models are commonly used for AQI prediction?"
response = query_engine.query(query)
print("🧠 Response from Query Engine:\n", response)


# 🧪 Step 8: Run Evaluations
faithfulness_result = faithfulness_evaluator.evaluate_response(
    query=query,
    response=response
)



# 📊 Step 9: Show Evaluation Results
print("\n📈 Evaluation Results")
print("✅ Faithful:", faithfulness_result.passing)
print("🔍 Faithfulness Feedback:", faithfulness_result.feedback)

```
